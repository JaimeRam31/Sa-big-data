```python

#importando algunas bibliotecas requeridas

import numpy as np 

import pandas as pd 

import sklearn

from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix 

from sklearn.tree import DecisionTreeClassifier 

from sklearn.metrics import accuracy_score 

from sklearn.datasets import load_iris

from matplotlib import pyplot as plt

from sklearn import datasets

from sklearn import tree

#Cargando conjuntos de datos

iris_data = load_iris() 

iris=pd.DataFrame(iris_data.data)

#características de impresión nombre de los datos del iris

print ("Features Name : ", iris_data.feature_names) 

#forma de conjuntos de datos

print ("Dataset Shape: ", iris.shape) 

#primera muestra de cinco

print ("Dataset: ",iris.head())  
```

```python
#impresión de muestras y destino

X = iris.values[:, 0:5] 

Y = iris_data.target

print(X)

print(Y)
```

```python
# Dividir el conjunto de datos en entrenar y probar

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

```python

# Clasificador de árboles de decisión

clf= DecisionTreeClassifier(random_state = 100)

#ajuste de los datos de entrenamiento

clf.fit(X_train, y_train)
```

```python
# predicción sobre datos aleatorios

X=[[6.4,1.8 ,6.6 ,2.1]]

Y_pred=clf.predict(X)

print(Y_pred)

# predicción en X_test (datos de prueba)

Y_pred=clf.predict(X_test)

print(Y_pred)


```

```python

#Accuray del modelo

print("Accuracy:", accuracy_score(y_test, Y_pred))

from sklearn.metrics import  confusion_matrix

cm=np.array(confusion_matrix(y_test,Y_pred))

cm
```

```python
#árbol trama

from sklearn.tree import plot_tree
plt.figure(figsize=(20, 10))

#tree.plot_tree(clf)
f = ['Longitud del sépalo (cm)', 'Ancho del sépalo (cm)', 'Longitud del pétalo (cm)', 'Ancho del pétalo (cm)']
s = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
plot_tree(clf, feature_names = f, class_names = s, rounded = True, filled= True)
plt.show()
```
